```python
import matplotlib.pyplot as plt
import os
import numpy as np
import nibabel as nib
#import cv2
import yaml
```

```python
with open('config.yaml') as file:
    config = yaml.load(file, Loader=yaml.FullLoader)
```

```python
config
```

# ACDC

```python
patient = '008'
frame = '01'
```

```python
img = nib.load(config['acdc']['path_acdc'] + '/patient' + patient + "/patient" + patient + '_frame' + frame + '.nii.gz')
msk = nib.load(config['acdc']['path_acdc'] + '/patient' + patient + "/patient" + patient + '_frame' + frame + '_gt.nii.gz')
```

```python
img = img.get_fdata()
msk = msk.get_fdata()
```

```python
img.shape, msk.shape
```

```python
img = img[:,:,0]
msk = msk[:,:,0]
```

```python
plt.figure(figsize=(10,10))

plt.subplot(121)
plt.title('acdc image')
plt.imshow(img, cmap = 'gray')
plt.axis('off')
plt.subplot(122)
plt.title('acdc mask')
plt.imshow(msk, cmap = 'gray')
plt.axis('off')

plt.axis('off')
plt.tight_layout()
```

```python
!mkdir data
```

```python
#cv2.imwrite('data/acdc.jpg', img)
#cv2.imwrite('data/acdc_gt.jpg', msk)
plt.imsave('data/acdc.jpg', img, format='jpeg', cmap='gray')
plt.imsave('data/acdc_gt.jpg', msk, format='jpeg', cmap='gray')
```
